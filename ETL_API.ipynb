```python
import pandas as pd
import sqlite3
```

```python
# Explore dataset
df = pd.read_csv('Data/Netflix_Userbase.csv')
```

```python
df
```

```python
df.info()
```

```python
# Connect to the sqlite db and a cursor
con = sqlite3.connect('NetflixUser.db')
cur = con.cursor()
```

```python
# Using pandas write the dataframes to tables
cur.execute("DROP TABLE IF EXISTS NetflixUser;")

df.to_sql('NetflixUser', con, index=False)
```

```python
# Run queries to test data
cur.execute("SELECT * FROM NetflixUser")
for row in cur:
    print(row)
```

```python
# Execute the query to select all rows from the table
cur.execute("SELECT * FROM NetflixUser")

# Fetch the column names
column_names = [desc[0] for desc in cur.description]

# Print the column names
for name in column_names:
    print(name)
```
